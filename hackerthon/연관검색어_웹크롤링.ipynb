```python
# 행정동 불러오기 

with open('towns.csv','r') as f:
    file = f.readlines()
    
    town_list = []
    
    for line in file:
        town_list.append(line[:-1])
        

#모듈 불러오기 

from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import quote

# url 인코더 

encoded_list = []
for town in town_list:
    encoded_list.append(quote(town))



#연관검색어 추출 

final_result = []

for town in town_list:
    town_encoded = quote(town)

    url = 'https://m.search.naver.com/search.naver?query={}&where=m&sm=mtp_hty'.format(town_encoded)
    web_page = BeautifulSoup(urlopen(url),"html5lib")

    box = web_page.find_all('a')

    each_result = []
    
    for i in range(58, 90):
        if box[i].get_text() == '신고':
            break
        else:
            each_result.append(box[i].get_text())
        
        
            
    
    final_result.append(each_result)
    
print(final_result)

```

```python
# counter 추가 

for n in range(len(final_result)):
    final_result[n].append([0,0,0])
    
print(final_result)
    
```
