```python
#모듈 불러오기 

from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import quote
```

```python
# 웹 크롤링 

url = 'http://www.seoul.go.kr/v2012/seoul/990908.html#a-01'
web_page = BeautifulSoup(urlopen(url),"html5lib")

table = web_page.find('div',{'id':'cont'})
name = table.find_all('td')


result_dong = []
result_dong2 = []
result = []

for t in range(len(name)):
    if t % 2 == 0:
        result_dong.append(name[t].get_text())
    else:
        result_dong2.append(name[t].get_text())

for dong in range(len(result_dong2)):
    result_dong22 = result_dong2[dong].split(',')
    result.append(result_dong22)

print(result)

```

```python
# 엑셀로 저장 

from openpyxl import Workbook

wb = Workbook()
ws1 = wb.active 
ws1.title = '행정동법정동'


for row in range(len(result)):
    for col in range(len(result[row])):
        ws1.cell(row=int(row)+1, column=int(col)+2, value=result[row][col])

for row in range(len(result_dong)):
    ws1.cell(row=int(row)+1, column=1, value=result_dong[row])
        
    
wb.save('C:/Users/User/test1.xlsx')
```
